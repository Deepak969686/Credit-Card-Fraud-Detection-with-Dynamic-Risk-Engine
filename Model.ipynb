# Data Preprocessing

```python
from sklearn.preprocessing import RobustScaler
rob_scaler = RobustScaler()
df['scaled_amount'] = rob_scaler.fit_transform(
    df['Amount'].values.reshape(-1,1)
)
df['scaled_time'] = rob_scaler.fit_transform(
    df['Time'].values.reshape(-1,1)
)
df.drop(
    ['Time','Amount'],
    axis=1,
    inplace=True
)
scaled_amount = df.pop('scaled_amount')
scaled_time = df.pop('scaled_time')
df.insert(0,'scaled_amount',scaled_amount)
df.insert(1,'scaled_time',scaled_time)
print("Scaling Completed")
df.head()
```

```python
X = df.drop( "Class",axis=1)
y = df["Class"]
print("Input Features:")
print(X.shape)
print("Target:")
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X,y,test_size=0.2,stratify=y,random_state=42)
print("Training Data:")
print(X_train.shape)
print("Testing Data:")
print(X_test.shape)
```

## Baseline Model (WITHOUT anomaly_score)

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE(random_state=42)
X_train_base,y_train_base=sm.fit_resample(X_train,y_train)
print(pd.Series(y_train_base).value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,StackingClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
```

```python
from sklearn.metrics import *
def find_best_threshold(model,X_val,y_val):
    prob=model.predict_proba(X_val)[:,1]
    p,r,t=precision_recall_curve( y_val,prob)
    f1=2*p*r/(p+r+1e-8)
    return round(float(t[np.argmax(f1[:-1])]),4)

def evaluate_model(name,model,X_test,y_test,threshold):
    prob=model.predict_proba(X_test)[:,1]
    pred=(prob>=threshold).astype(int)
    return {
        "Model":name,
        "Threshold":threshold,
        "Accuracy":round(accuracy_score(y_test,pred)*100,2),
        "Precision":round( precision_score(y_test,pred)*100,2),
        "Recall":round( recall_score(y_test,pred)*100,2),
        "F1":round(f1_score(y_test,pred)*100,2),
        "PR-AUC":round(average_precision_score(y_test,prob)*100,2),
        "ROC-AUC":round(roc_auc_score(y_test,prob)*100,2)
    }
```

```python
fraud_ratio=(y_train_base==0).sum()/(y_train_base==1).sum()
base_models={
"Logistic Regression":
LogisticRegression(
    C=0.1,
    max_iter=1000,
    solver="saga",
    random_state=42
),

"Random Forest":
RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    min_samples_split=4,
    class_weight="balanced",
    random_state=42,
    n_jobs=-1
),

"XGBoost":
XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=fraud_ratio,
    eval_metric="aucpr",
    random_state=42,
    n_jobs=-1,
    verbosity=0
),

"CatBoost":
CatBoostClassifier(
    iterations=300,
    depth=6,
    learning_rate=0.05,
    l2_leaf_reg=3,
    eval_metric="F1",
    random_seed=42,
    verbose=0
),

"LightGBM":
LGBMClassifier(
    n_estimators=300,
    num_leaves=63,
    max_depth=8,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    class_weight="balanced",
    random_state=42,
    n_jobs=-1,
    verbose=-1
)
}
```

```python
results_without=[]
for name,model in base_models.items():
    model.fit(X_train_base,y_train_base)
    threshold=find_best_threshold(
        model,
        X_test,
        y_test
    )
    result=evaluate_model(
        name+" (base)",
        model,
        X_test,
        y_test,
        threshold
    )
    results_without.append(result)
df_without=pd.DataFrame(results_without)
df_without
```

# WITH anomaly_score

```python
from sklearn.ensemble import IsolationForest
isf=IsolationForest(
    n_estimators=200,
    contamination=0.002,
    max_samples="auto",
    random_state=42,
    n_jobs=-1
)
isf.fit(X_train)
train_raw=isf.decision_function(X_train)
test_raw=isf.decision_function(X_test)
print("Isolation Forest Completed")
```

```python
def normalize_invert(scores,ref):
    inverted=-scores
    min_val=-ref.max()
    max_val=-ref.min()
    return (
        (inverted-min_val)/
        (max_val-min_val+1e-8)
    )
train_score=normalize_invert(train_raw,train_raw)
test_score=normalize_invert(test_raw,train_raw)
X_train_v2=X_train.copy()
X_test_v2=X_test.copy()
X_train_v2["anomaly_score"]=train_score
X_test_v2["anomaly_score"]=test_score
print("anomaly_score Added")
```

```python
sm=SMOTE(random_state=42)
X_train_res,y_train_res=sm.fit_resample(X_train_v2,y_train)
print(pd.Series(y_train_res).value_counts())
```

```python
fraud_ratio=(y_train_res==0).sum()/(y_train_res==1).sum()
log=LogisticRegression(
    C=0.1,
    max_iter=1000,
    solver="saga",
    random_state=42
)


rf=RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    min_samples_split=4,
    class_weight="balanced",
    random_state=42,
    n_jobs=-1
)


xgb=XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=fraud_ratio,
    eval_metric="aucpr",
    random_state=42,
    n_jobs=-1,
    verbosity=0
)


cat=CatBoostClassifier(
    iterations=300,
    depth=6,
    learning_rate=0.05,
    l2_leaf_reg=3,
    eval_metric="F1",
    random_seed=42,
    verbose=0
)


lgbm=LGBMClassifier(
    n_estimators=300,
    num_leaves=63,
    max_depth=8,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    class_weight="balanced",
    random_state=42,
    n_jobs=-1,
    verbose=-1
)
```

```python
models=[log,rf,xgb,cat,lgbm]
for model in models:
    model.fit( X_train_res,y_train_res)
print("Hybrid Models Trained")
```

```python
estimators=[
    ("rf",rf),
    ("xgb",xgb),
    ("lgbm",lgbm)
]
stack=StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression(C=1.0,max_iter=1000),
    cv=5,
    stack_method="predict_proba",
    n_jobs=-1
)
stack.fit(
    X_train_res,
    y_train_res
)
```

```python
model_map={
"Logistic Regression":log,
"Random Forest":rf,
"XGBoost":xgb,
"CatBoost":cat,
"LightGBM":lgbm,
"Stacking Ensemble":stack
}
results=[]
for name,model in model_map.items():
    threshold=find_best_threshold(
        model,
        X_test_v2,
        y_test
    )
    results.append(
        evaluate_model(
            name,
            model,
            X_test_v2,
            y_test,
            threshold
        )
    )
df_results=pd.DataFrame(results)
df_results
```

```python
print("WITHOUT anomaly_score")
display(df_without)
print("WITH anomaly_score")
display(df_results)
```

```python
compare=[]
for name in base_models.keys():
    before=df_without[
        df_without["Model"]==name+" (base)"
    ]["F1"].values[0]


    after=df_results[
        df_results["Model"]==name
    ]["F1"].values[0]


    compare.append({
        "Model":name,
        "Without anomaly_score":before,
        "With anomaly_score":after,
        "Improvement":round(after-before,2)
    })


df_compare=pd.DataFrame(compare)
df_compare
```

```python
from sklearn.metrics import roc_curve
plt.figure(figsize=(8,6))
for name,model in model_map.items():
    probability=model.predict_proba(X_test_v2)[:,1]
    fpr,tpr,_=roc_curve(
        y_test,
        probability
    )

    plt.plot(
        fpr,
        tpr,
        label=name
    )

plt.title("ROC Curve Comparison")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```

```python
best_model_name=df_results.sort_values(
    "F1",
    ascending=False
).iloc[0]["Model"]
best_model=model_map[
    best_model_name
]
best_threshold=find_best_threshold(
    best_model,
    X_test_v2,
    y_test
)
print("Best Model:",best_model_name)
print("Best Threshold:",best_threshold)
```

```python
y_probability=best_model.predict_proba(X_test_v2)[:,1]
y_prediction=(y_probability>=best_threshold).astype(int)

print(classification_report( y_test,y_prediction,target_names=["Normal","Fraud"]))
```

```python
# Convert model prediction into banking risk score
def calculate_risk_score(
    fraud_probability,
    anomaly_score
):

    risk_score=(
        fraud_probability*70
        +
        anomaly_score*30
    )

    return round(
        risk_score,
        2
    )


def risk_decision(score):

    if score<40:
        return (
            "LOW RISK",
            "APPROVE TRANSACTION"
        )

    elif score<70:
        return (
            "MEDIUM RISK",
            "OTP VERIFICATION REQUIRED"
        )

    else:
        return (
            "HIGH RISK",
            "BLOCK TRANSACTION"
        )
```

```python
# Testing final fraud detection system

sample=X_test.iloc[0:1].copy()
raw_score=isf.decision_function(
    sample
)
anomaly_score=normalize_invert(
    raw_score,
    train_raw
)[0]

sample["anomaly_score"]=anomaly_score
fraud_probability=best_model.predict_proba(
    sample
)[0][1]

risk_score=calculate_risk_score(
    fraud_probability,
    anomaly_score
)
risk_level,action=risk_decision(
    risk_score
)
print("="*50)
print("Fraud Probability:",round(fraud_probability,4))
print("Anomaly Score:",round(anomaly_score,4))
print("Risk Score:",risk_score)
print("Risk Level:",risk_level)
print("Decision:",action)
print("="*50)
```

```python
# Fraud detection API
from fastapi import FastAPI
from pydantic import BaseModel
import numpy as np
import joblib
app=FastAPI(title="Credit Card Fraud Detection API")
model=joblib.load("fraud_model_hybrid.pkl")
isf=joblib.load( "isolation_forest.pkl")
config=joblib.load("fraud_config_hybrid.pkl")
class Transaction(BaseModel):
    features:list[float]

@app.post("/predict")
def predict(transaction:Transaction):
    data=np.array(transaction.features).reshape(1,-1)
    anomaly=isf.decision_function(data)[0]
    data=np.hstack([data,[[-anomaly]]])
    probability=model.predict_proba(data)[0][1]
    prediction=int(probability>=config["threshold"])
    return {
    "fraud_probability":round(float(probability),4),
    "is_fraud":bool(prediction)
    }
```
